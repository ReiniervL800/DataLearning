```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (15,8))
plt.xticks(rotation=45)
sns.violinplot(x='Huishoudkenmerken',y='TevredenheidMetDeHuidigeWoning_1',data=df)
plt.title('Tevredenheid woning')
plt.xlabel('Huishoud kenmerken')
plt.ylabel('Tevredenheid')
plt.show()
```
